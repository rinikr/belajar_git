```python
#Mencari perbandingan antara jumlah pelanggan yang churn dan tidak churn, mencari rata-rata berdasar monthlycharges dan tenure
#filter yang churn
df1[(df1['Churn']=="Yes")]
```

```python
#jumlah pelanggan yang churn 1869 orang, sedang yang tidak churn  5163 orang
```

```python
#Mencari perbandingan rata-rata yang churn dan tidak berdasar monthlycharges
df.groupby("Churn").agg({
    "MonthlyCharges":["mean"]})
```

```python
#Mencari perbandingan rata-rata yang churn dan tidak berdasar tenure
df.groupby("Churn").agg({
    "tenure":["mean"]})
```
